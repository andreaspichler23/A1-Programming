```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
nmax = 1000
r2score_list = []

for i in range(10,nmax):   
    x_list = np.random.randn(i)
    y_fit_list = [ x**2 for x in x_list ]
    y_data_list = [ y + np.random.randn() for y in y_fit_list ]

    r2score_list.append(r2_score(y_data_list, y_fit_list))
 

#plt.scatter(x_list, y_data_list)
plt.plot(r2score_list)
    
```
