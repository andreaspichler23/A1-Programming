```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
nmax = 1000
iterations = 10
error_list_final = []
regressor = linear_model.LinearRegression()

for i in range(10,nmax):
    error_list = []
    for j in range(iterations):
        
        x_list = np.random.randn(i).reshape(-1,1)
        y_list_true = [ 2*x for x in x_list ]
        y_data_list = [ y + np.random.randn() for y in y_list_true ]

        regressor.fit( x_list, y_data_list )    
        y_list_pred = regressor.predict(x_list)
        
        error_list.append( mean_squared_error( y_list_true, y_list_pred  ) )
    
    error_list_final.append( np.mean(error_list) * np.sqrt(i) )
 

#plt.scatter(x_list, y_data_list)
plt.plot(error_list_final)
    
```
